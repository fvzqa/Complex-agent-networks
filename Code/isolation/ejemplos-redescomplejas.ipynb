```python
import networkx as nx
import pandas as pd
import numpy as np
import random
from matplotlib import pyplot as plt
```

```python
G = nx.erdos_renyi_graph(25,0.4)
```

```python
nx.draw(G)
plt.savefig('random.png', dpi=300, bbox_inches='tight')
plt.show()
```

```python
H = nx.scale_free_graph(25)
```

```python
nx.draw(H)
```

```python
H = H.to_undirected()
```

```python
nx.draw(H)
```

```python
H.remove_edges_from(nx.selfloop_edges(H))
```

```python
nx.draw(H)
plt.savefig('small-world.png', dpi=300, bbox_inches='tight')
plt.show()
```

```python
F = nx.navigable_small_world_graph(5)
```

```python
nx.draw(F)
```

```python
H = F.remove_edges_from(nx.selfloop_edges(F))
F = F.to_undirected()
nx.draw(F)
plt.savefig('small-world.png', dpi=300, bbox_inches='tight')
plt.show()
```
