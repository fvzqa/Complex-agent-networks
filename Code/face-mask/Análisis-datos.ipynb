```r
library(ggplot2)
```

```r
data <- data.frame(c(rep(0,5000),rep(10,5000),rep(20,5000), rep(30,5000),rep(40,5000),rep(50,5000),rep(60,5000), rep(70,5000), 
                     rep(80,5000), rep(90,5000), rep(100,5000)))

x <- c(x0,x1,x2,x3,x4,x5,x6,x7,x8,x9,x10)

data <-cbind(data,x)

names(data) <- c('Porcentaje', 'Maxinfectados')

data$Porcentaje <- as.factor(data$Porcentaje)

png("face-mask(1).png", width = 2500, height = 1500)
p <- ggplot(data, aes(x=Porcentaje, y=Maxinfectados)) + 
  geom_violin(aes(fill= Porcentaje), scale = "width", colour = "black")
p + geom_boxplot(width=0.2, fill = "slategray", colour = "black")+
    theme(
  panel.background = element_rect(fill = "white", 
                                size = 2, linetype = "solid"),
  panel.grid.major = element_line(size = 0.5, linetype = 'solid',
                                colour = "gray"), 
  panel.grid.minor = element_line(size = 0.25, linetype = 'solid',
                                colour = "gray")
  ) + theme(axis.title.y = element_text(size = 50, angle = 90))+
    theme(axis.text.x = element_text(angle = 00, hjust = 1, size=35,color="black"))+
theme(axis.text.y = element_text(angle = 00, hjust = 1, size=35,color="black")) +
theme(axis.title.x = element_text(size = 50, angle = 00)) + xlab("Porcentaje de la población que usa cubrebocas") + ylab("Cantidad de Infectados") + theme(legend.position="none")+
scale_fill_manual(values = c("#BCE4D8","#BCE4D8","#BCE4D8","#BCE4D8","#BCE4D8","#BCE4D8","#BCE4D8","#BCE4D8","#BCE4D8","#BCE4D8","#BCE4D8"))


dev.off()
```

```r
p <- ggplot(data, aes(x = Porcentaje, y = Maxinfectados, fill = Porcentaje)) + geom_violin(trim = FALSE) + geom_boxplot(width = 0.07) + scale_fill_brewer() + theme(legend.position = "none")
p
```
