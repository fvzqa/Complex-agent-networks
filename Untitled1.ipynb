```r
z <- read.csv("network_data.csv")

names(z) <- c("index", "time", "susc", "inf", "rec")
```

```r
plot(z[c(2,3)], col="firebrick3", ylim=c(0, 1000), main="", 
     ylab="Cantidad de individuos", xlab="tiempo")
points(z[c(2,4)], col="darkgoldenrod3")
points(z[c(2,5)], col="darkblue")
legend(x = "right", legend = c("Susceptible", "Infeccioso", "Removido"), 
       fill = c("firebrick3", "darkgoldenrod3","darkblue"), 
       title = "")
```
